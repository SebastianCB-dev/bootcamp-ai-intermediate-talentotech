```python
#open cv
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
rutaimagen=cv.imread('./roya.jpg')
#utiliza el formtao BGR, se debe cambiar a RGB
imagen=cv.cvtColor(rutaimagen,cv.COLOR_BGR2RGB)
plt.imshow(imagen)
plt.show()
```

```python
#se cambia el formato a HSV
hsv=cv.cvtColor(imagen,cv.COLOR_RGB2HSV)
plt.imshow(hsv)
plt.show()
```

```python
#se define un rango de naranja para detectar Roya
azulbajo=np.array([10,90,110])
azulalto=np.array([18,255,255])
mascara=cv.inRange(hsv,azulbajo,azulalto)
plt.imshow(mascara)
plt.show()
```

```python
#selecciona pixeles de la imagen original
res=cv.bitwise_and(imagen,imagen,mask=mascara)
plt.imshow(res)
plt.show()
```
