```python
# Víctor Viera Balanta
#se carga las bibliotecas
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn import datasets
```

```python
#sklearn trae la data de iris
datos=  datasets.load_iris()
datos
```

```python
#se selecciona las medidas, todas las filas, selecciona desde la coloumna cero, hasta la tres
X = datos['data'][:, 0:3]
#los labels
y = datos['target']
```

```python

#se dividen los datos en prueba y  entrenamiento
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 82)
```

```python
#transformación de escala de los datos
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#se llama al modelo de Naive Gaussiano
from sklearn.naive_bayes import GaussianNB
nvclassifier = GaussianNB()
#se entrena el modelo, Probabilidad, modelo probabilidad condicionada
nvclassifier.fit(X_train, y_train)
```

```python
# se realiza la paredicción, con el entrenamiento
y_pred = nvclassifier.predict(X_train)
print(y_pred)
```

```python
# Métrica de la matriz de confusión
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_train, y_pred)
print(cm)
```

```python
#reporte de clasificación
print(classification_report(y_train, y_pred))
```

```python
# se realiza la predicción con los datos de prueba
y_predTest = nvclassifier.predict(X_test)
print(y_predTest)
```

```python
#matriz de confusión
cm = confusion_matrix(y_test, y_predTest)
print(cm)
```

```python
#reporte de clasificación
print(classification_report(y_test, y_predTest))
```
