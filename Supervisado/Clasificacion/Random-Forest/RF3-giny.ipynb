# Random Forest

Este es un ejemplo usando el algoritmo Random Forest usando la librería scikit-learn y el metodo de Giny impureza.

```python
# Importación de librerías
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Importación de datos
datos = pd.read_csv('./assets/dataset2.csv', delimiter=',')
datos.head()
```

```python
# Describir datos
datos.describe()
```

```python
# Información de los datos
datos.info()
```

```python
# Correlación de los datos
datos.corr()
```

```python
# Preparar los datos
X = datos.drop(columns=['Loves movie'])
y = datos['Loves movie']
```

```python
# Modelo de Random Forest con el uso de entropía
clf = tree.DecisionTreeClassifier(max_leaf_nodes=4, random_state=0, criterion='gini')
clf = clf.fit(X, y)
```

```python
# Imprimir el arbol
tree.plot_tree(clf)
```

```python
# Imprimir arbol con nombres de columnas
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20, 10))
tree.plot_tree(clf, feature_names=X.columns, class_names=['No', 'Si'], filled=True, ax=ax)
plt.show()
```
