```python
# Víctor Viera Balanta
#cargan las bibliotecas
from sklearn import datasets
```

```python
datos=  datasets.load_digits()
datos
```

```python
#X = datos['feature_names']
X = datos.data
y = datos['target']
```

```python

#se dividen los datos en prueba y entrenamiento
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
#se transforman los datos en scala
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#se transforman los a una escala estándar
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python

from sklearn.naive_bayes import GaussianNB
nvclassifier = GaussianNB()
#se realiza en entrenamiento
nvclassifier.fit(X_train, y_train)
```

```python
#las prediccione
y_pred = nvclassifier.predict(X_train)
print(y_pred)
```

```python
#matriz de confusión de Naive Bayes
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_train, y_pred)
print(cm)
```

```python
print(classification_report(y_train, y_pred))
```

```python

y_predTest = nvclassifier.predict(X_test)
print(y_predTest)
```

```python
cm = confusion_matrix(y_test, y_predTest)
print(cm)
```

```python
print(classification_report(y_test, y_predTest))
```
