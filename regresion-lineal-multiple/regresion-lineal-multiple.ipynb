```python
# Importación de librerías
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Leer dataset
df = pd.read_csv('./assets/FuelConsumptionCo2.csv')
df.head() # Mostrar los primeros 5 registros
```

```python
# Tomar columnas numericas
df = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
df.head() # Mostrar los primeros 5 registros
```

```python
# Ver correlación
df.corr()
```

```python
# Columnas a tomar (FUELCONSUMPTION_CITY, FUELCONSUMPTION_COMB, FUELCONSUMPTION_COMB_MPG, CO2EMISSIONS)
df = df[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
df.head() # Mostrar los primeros 5 registros
```

```python
# Total de ejemplos
total_examples = df.shape[0]
total_examples
```

```python
# Dividir conjunto de datos
x_train, x_test, y_train, y_test = train_test_split(df[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG']], df['CO2EMISSIONS'], test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Entrenar modelo
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Evaluar modelo
model.score(x_test, y_test)
```
