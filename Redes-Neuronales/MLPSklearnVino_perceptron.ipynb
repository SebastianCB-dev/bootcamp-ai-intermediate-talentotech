```
X = wine.drop('Wine',axis=1)
y = wine['Wine']
```

```
X
```

```
X
```

```
# se llama para partir los datos de prueba y entrenamiento
from sklearn.model_selection import train_test_split
```

```
#se dividen los datos
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```
#calcula los patrones para estandarizar los datos
scaler.fit(X_train)
```

```
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```
from sklearn.neural_network import MLPClassifier
#configuro el número de capas con sus neurones
mlp = MLPClassifier(hidden_layer_sizes=(13,13,13),max_iter=500)
```

```
#backpropagation, entreno el modelo
mlp.fit(X_train,y_train)
```

```
#se realiza la predicción , prueba
predictions = mlp.predict(X_test)
predictions
```

```
from sklearn.metrics import classification_report,confusion_matrix
#matriz de confusión
print(confusion_matrix(y_test,predictions))
cmatrix=confusion_matrix(y_test,predictions)
```

```
cmPd= pd.DataFrame(cmatrix)
```

```
#Graficamos la matriz de confusión
import seaborn as sn
import matplotlib.pyplot as plt
sn.set(font_scale=1.4) # for label size
sn.heatmap(cmPd, annot=True, annot_kws={"size": 20},fmt='g',center=0,linewidths=1,cbar=False) # font size
plt.show()
```

```
#se convierten a data Frame con pandas
predictionsdp=pd.DataFrame(predictions)
#se grafican con seaborn
import seaborn as sns
import matplotlib.pyplot as plt
#https://seaborn.pydata.org/generated/seaborn.pairplot.html
#se crea la gráfica tipo kde(curvano),kidn=scatter-dispersión
sns.pairplot(predictionsdp,hue = None,diag_kind = "kde",kind = "scatter",palette = "husl",height=3.5)
plt.show()
```

```
print(classification_report(y_test,predictions))
```
